```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv")
df
```

```python
# df['Date'] = pd.to_datetime(df['Date'])
df.sort_values(by=['ID', 'Date'], inplace = True)

# df['Year'] = df['Date'[:4]]
df['Year'] = df['Date'].apply(lambda x: str(x[:4]))
df['Month-Day'] = df['Date'].apply(lambda x: str(x[5:]))
df = df[df['Month-Day'] != '02-29']
df
```

```python
df_temp_min_2014 = df[(df['Element'] == 'TMIN') & (df['Year'] != '2015')]
df_temp_max_2014 = df[(df['Element'] == 'TMAX') & (df['Year'] != '2015')]

df_temp_min_2015 = df[(df['Element'] == 'TMIN') & (df['Year'] == '2015')]
df_temp_max_2015 = df[(df['Element'] == 'TMAX') & (df['Year'] == '2015')]
```

```python
import numpy as np
df_min_2014 = df_temp_min_2014.groupby(['Month-Day'])['Data_Value'].agg({'Min Temp': np.min})
df_max_2014 = df_temp_max_2014.groupby(['Month-Day'])['Data_Value'].agg({'Max Temp': np.max})
df_max_2014
```

```python
df_min_2015 = df_temp_min_2015.groupby(['Month-Day'])['Data_Value'].agg({'Min Temp': np.min})
df_max_2015 = df_temp_max_2015.groupby(['Month-Day'])['Data_Value'].agg({'Max Temp': np.max})
df_max_2015
```

```python
df_max_2014.reset_index(inplace = True)
df_max_2015.reset_index(inplace = True)
df_min_2015.reset_index(inplace = True)
df_min_2014.reset_index(inplace = True)
df_min_broken_index = df_min_2015[df_min_2015['Min Temp'] < df_min_2014['Min Temp']].index.tolist()
df_min_broken_index
```

```python
df_max_broken_index = df_max_2015[df_max_2015['Max Temp'] > df_max_2014['Max Temp']].index.tolist()
df_max_broken_index
```

```python
plt.plot(df_max_2014['Max Temp'], label = 'Record Max Temperature')
plt.plot(df_min_2014['Min Temp'], label = 'Record Min Temperature')
plt.scatter(df_min_broken_index, df_min_2015['Min Temp'].iloc[df_min_broken_index], label = 'Min Temp Broken in 2015', s = 3, c = 'g')
plt.scatter(df_max_broken_index, df_max_2015['Max Temp'].iloc[df_max_broken_index], label = 'Max Temp Broken in 2015', s = 3, c = 'r')
plt.gca().fill_between(range(len(df_max_2014)), df_max_2014['Max Temp'], df_min_2014['Min Temp'], alpha = 0.2)
# plt.show()

plt.xlabel('Month')
plt.ylabel('Temperature (Tenths of Degrees C)')
plt.title('Extreme Temperatures of 2015 against 2005-2014\n Ann Harbour, Michigan, USA')

plt.legend(frameon = False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().axis([0,365,-400,400])

a = [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330]
b = [i+15 for i in a]

Month_name = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(b, Month_name)
```
